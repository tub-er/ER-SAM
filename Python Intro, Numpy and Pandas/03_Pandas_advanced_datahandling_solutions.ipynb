### Do you remember how to filter a DataFrame?

Filter for:
* Installed capacity >= 2000 kW
* Energy carrier: wind (onshore/offshore)
* Energy carrier: contains the word 'gas'

```python
# filter for installed capacity over 2000 kW
dataset[dataset['4.2 Installierte Leistung [kW]'] > 2000]
```

```python
# filter for energy carrier is wind
dataset[(dataset['4.1 Energieträger'] == 'Wind Land') | 
        (dataset['4.1 Energieträger'] == 'Wind See')]
```

```python
# filter for energy carrier contains the word gas
dataset[dataset['4.1 Energieträger'].str.contains('gas')]
```

#### Task: use the groupby function to filter for:
- installed capacity per federal state
- installed capacity per energy carrier 

```python
# get installed capacity per federal state
dataset.groupby('4.11 Bundesland')['4.2 Installierte Leistung [kW]'].sum()
```

```python
# plot
dataset.groupby('4.11 Bundesland')[
    '4.2 Installierte Leistung [kW]'
].sum().plot.barh(xlabel='kW')
```

```python
# get installed capacity per energy carrier
dataset.groupby('4.1 Energieträger')['4.2 Installierte Leistung [kW]'].sum()
```

```python
# plot
dataset.groupby('4.1 Energieträger')[
    '4.2 Installierte Leistung [kW]'
].sum().plot.barh(xlabel='kW')
```

#### Task: Use the pivot_table function to clean up the dataframe and retain relevant information only:

- Installed capacity per federal state and energy carrier


```python
pivot_df = pd.pivot_table(dataset, 
               index='4.11 Bundesland', 
               columns='4.1 Energieträger', 
               values='4.2 Installierte Leistung [kW]',
               aggfunc='sum'
              )
pivot_df
```

```python
pivot_df.plot.barh(stacked=True)
```
